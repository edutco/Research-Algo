```python
import sqlite3
import requests
import xmltodict
db = sqlite3.connect('my_database2.db')
cursor = db.cursor()
cursor.execute('''
    CREATE TABLE IF NOT EXISTS KNS_BillUnion(
        BillUnionID     INTEGER   PRIMARY KEY, 
        MainBillID   INTEGER,
        UnionBillID  INTEGER, 
        LastUpdatedDate DATETIME2)
''')
response=requests.get('http://knesset.gov.il/Odata/ParliamentInfo.svc/KNS_BillUnion')#?$filter=KnessetNum eq 20')
dict_data=xmltodict.parse(response.content)
for line in dict_data['feed']['entry']:
        data = line['content']['m:properties']
        BillUnionID = data['d:BillUnionID']['#text']
        MainBillID = data['d:MainBillID']['#text']
        UnionBillID = data['d:UnionBillID']['#text']
        LastUpdatedDate = data['d:LastUpdatedDate']['#text']
        #print(BillUnionID,MainBillID,UnionBillID,LastUpdatedDate)
        params = [BillUnionID, MainBillID, UnionBillID, LastUpdatedDate]
        cursor.executemany('''INSERT OR IGNORE INTO KNS_BillUnion(BillUnionID, MainBillID, UnionBillID, LastUpdatedDate)
         VALUES (?,?,?,?)''', [params])
        db.commit()
cursor.execute('''SELECT * FROM KNS_BillUnion''')
all_rows = cursor.fetchall()
print()
for row in all_rows:
    print('{0} : {1}, {2}, {3}'.format(row[0], row[1], row[2], row[3]))   
cursor.execute("DROP TABLE KNS_BillUnion")                    
```
